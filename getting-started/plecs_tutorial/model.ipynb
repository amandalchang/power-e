# Getting PLECS Simulation Data Into Your Report

## PLECS Schematic

To copy a PLECS Schematic, select "Edit -> Copy as Image". Then, simply paste into a Jupyter markup cell. Jupyter will encode the bitmap directly into the cell.





## PLECS Scope Plots

The following PLECS Scope plot is also copy-pasted as an image. With the scope plot active, select "Edit->Copy". PLECS will offer several formatting options. Then, paste into the Jupyter markup cell.






It is also possible to export PLECS Scope data as a CSV file. With the PLECS Scope active, select "Export -> As CSV -> All..." and save as e.g. "data.csv".

You can load and inspect a CSV file as follows:

```python
data = pd.read_csv('data.csv')
print(data.head())
print("\n")

data_new = pd.read_csv('data_new.csv')
print(data_new.head())
```

Plotting and/or processing the data is simple once the names of the signals.

```python
data.plot(x="Time / s", y="Sine Wave")

data.plot(x="Time / s", y="Random Numbers")
```

```python
# rename a trace
data.rename(columns = {"Random Numbers":"Noise"}, inplace = True)

# filter the data
data['Noise filtered'] = data["Noise"].rolling(100).mean()

data.plot(x="Time / s", y=["Noise", "Noise filtered"])
```
