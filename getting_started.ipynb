### First Explorations
#### Please navigate the map of Power Electronics and answer the following questions (note: if the map does not load correctly, use the Canvas App or try a different browser):
* Provide an example for a mixed-signal circuit.
  > A gate driver circuit compatible with PWM control that can level-shift.
* What type of quantities are typically measured for the control and protection of power converters?
  > Current, voltage, and temperature.
* What is the "loop" in loop shaping?
  > The loop in loop shaping refers to the control loop, as defined by a transfer function, of a closed-loop system.
* How does mechanical engineering apply to the design of power converters?
  > Packaging is a huge part of power electronics and is very important to the success of the electronic, especially in regards to wicking away heat. Other very important mechanical design considerations include vibration, EMI, thermal, and space constraints.
* What purpose do inductors serve in power converters?
  > Inductors store energy in their magnetic field to filter the high frequency harmonics that come from the voltage modulation away.
* What is currently "all the rage" in the field of power electronics?
  > Wide-bandgap devices, which are capable of operating at higher temperatures with much less switching loss than conventional silicon devices.
* Who is James Worden?
  > James Worden is the co-founder of Solectria, a company which makes powerful large scale high-voltage inverters.
* Why does the efficiency of a power converter matter?
  >  A more efficient power converter is a more reliable and cooler converter. High-power converters need to be especially efficient or else even 1% losses can result in huge amounts of heat.
* Which are some of the key powerstage components?
  > Power semiconductors (like MOSFETs and diodes), magnetics (i.e. inductors and transformers), capacitors, and protective components.
* List multiple ways of getting hurt or killed when working on a power converter.
  > You can get shocked, burned, or hit with debris.
* How do you plan on staying safe.
  > Wearing safety glasses and not touching the lab hardware while it's on.
